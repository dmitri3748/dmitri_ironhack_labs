```python
# group by the ones who answered with tca
responses_by_tca = df.groupby(['response', 'total_claim_amount'])['response'].count().rename('count').reset_index()
```

```python
responses_by_tca
# there seems to be a big difference between the ones who answered and the ones who have not answered
```

```python
# would it be interesting to have two boxplots to see how they compare?
```

```python
fig1, axes1 = plt.subplots(figsize = (16,6))

sns.barplot(x='total_claim_amount', y ='count', hue = 'response', data = responses_by_tca)
```

```python
yes_mean = responses_by_tca[responses_by_tca['response']=='yes']['total_claim_amount'].mean()
no_mean = responses_by_tca[responses_by_tca['response']=='no']['total_claim_amount'].mean()


sns.boxplot(data=responses_by_tca, x='response', y='total_claim_amount')
plt.plot([0, 1], [yes_mean, no_mean], marker='o', linestyle='--', color='red')


plt.title('Distribution of Claim Amount by Response')
plt.xlabel('Response')
plt.ylabel('Claim Amount')

plt.show()
```

```python
responses_by_income = df.groupby(['response', 'income'])['response'].count().rename('count').reset_index()
```

```python
yes_mean = responses_by_income[responses_by_income['response']=='yes']['income'].mean()
# need to change the groupby
no_mean = responses_by_income[responses_by_income['response']=='no']['income'].mean()


sns.boxplot(data=responses_by_income, x='response', y='income')
plt.plot([0, 1], [yes_mean, no_mean], marker='o', linestyle='--', color='red')


plt.title('Distribution of Income by Response')
plt.xlabel('Response')
plt.ylabel('Income')

plt.show()
```
