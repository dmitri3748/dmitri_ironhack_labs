```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

men = tips.tip.loc[(tips['sex'] == 'Male')]
women = tips.tip.loc[(tips['sex'] == 'Female')]

# Create a figure with a single subplot
fig, ax = plt.subplots(figsize=(8, 4))
 
# Plot the KDEs for men and women in the same subplot
sns.kdeplot(data=men, ax=ax, color='blue', shade=True, label='Men')
sns.kdeplot(data=women, ax=ax, color='orange', shade=True, label='Women')

# Set the title and axis labels for the subplot
ax.set_title('Distribution of Tips by Gender')
ax.set_xlabel('Tip Amount')
ax.set_ylabel('Density')

# Add a legend for the KDEs
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate some random data
np.random.seed(42)
men = np.random.normal(5, 2, size=100)
women = np.random.normal(7, 1, size=100)

# Create a figure with two subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Plot the KDEs on the subplots
sns.kdeplot(data=men, ax=axs[0], color='blue', shade=True)
sns.kdeplot(data=women, ax=axs[1], color='orange', shade=True)

# Adjust the position of the second subplot to overlap the first one
pos = axs[0].get_position()
axs[1].set_position([pos.x0 + 0.2, pos.y0, pos.width, pos.height])

# Set the titles for the subplots
axs[0].set_title('Men')
axs[1].set_title('Women')

# Remove the y-axis labels from the second subplot
axs[1].set_ylabel('')

# Add a common x-axis label
fig.text(0.5, -0.05, 'Tip Amount', ha='center')

# Show the plot
plt.show()
```
