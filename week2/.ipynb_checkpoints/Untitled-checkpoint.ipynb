```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titatic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips.head()
```

```python
plt.figure(figsize=(8, 6))


sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', size = 'size', \
                palette = 'YlGnBu', alpha = 0.8, style = 'sex', markers=['o', 'o']). \
                set_title('Relationship between Total Bill and Tips by Gender and Food Size') 

# adjust axis labels
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')


# annotate the outlier
plt.annotate('Outlier', xy=(50, 10), xytext=(70, 10), arrowprops=dict(facecolor='black', shrink=0.01))

# palette='RdPu'
```

```python
sns.histplot(tips.tip, kde = True, bins = 100)
```

```python
sns.displot(tips.tip, kde = True)
```

```python
sns.barplot(data = tips, x = 'sex', y = 'tip')
```

```python
fig = plt.gcf()
fig.set_facecolor('lightgray')

colors = {'Male': 'red', 'Female': 'white'}
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', \
            hue_order=['Male', 'Female'], palette=colors).set_facecolor('lightblue')
```

```python
sns.stripplot(data = tips, x = 'day', y = 'tip', hue = 'sex', dodge = True)
```
